```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = {
    'Fruit': ['durian', 'papaya', 'lychee', 'mango'],
    'Probability': [0.05, 0.15, 0.3, 0.5]
}
df = pd.DataFrame(data)

sns.set_style("dark")

cmap = sns.dark_palette("blue", as_cmap=True)
colors = [cmap(val) for val in df['Probability']]

fig, ax = plt.subplots(figsize=(8, 6), facecolor='black')
ax.set_facecolor("black")

bars = ax.barh(df['Fruit'], df['Probability'], color=colors, edgecolor='white')

for bar, prob in zip(bars, df['Probability']):
    ax.text(prob + 0.02, bar.get_y() + bar.get_height() / 2, f"{prob:.2f}", 
            va='center', ha='left', color='white')

ax.set_xlim(0, 1)
ax.set_xlabel('Probability', color='white', fontsize=12)
ax.set_ylabel('word', color='white', fontsize=12)
ax.tick_params(colors='white')

plt.tight_layout()
plt.show()
```
